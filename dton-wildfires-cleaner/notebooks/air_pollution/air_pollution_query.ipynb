```python
from utils import get_env_var, print_titles
from event_registry_utils import EventRegistry
from events_utils import filter_events_by_concept_thresholds
```

```python
EVENT_REGISTRY_API_KEY = get_env_var("EVENT_REGISTRY_API_KEY")
```

```python
event_registry = EventRegistry(EVENT_REGISTRY_API_KEY, use_cache=True)
```

```python
date_start = "2020-01-01"
date_end = "2020-12-31"
```

Things I tried to improve:
    - exclude {"categoryUri": "dmoz/Shopping..."} --> eliminated a lot of real events
    - leaving air quality index out returns way to many results 

```python
query1 = {
    "$query": {
        "$and": [
            {"conceptUri": "http://en.wikipedia.org/wiki/Air_pollution"},
            {"conceptUri": "http://en.wikipedia.org/wiki/Air_quality_index"},
            {"categoryUri": "dmoz/Science/Environment/Air_Quality"},
            {
                "dateStart": date_start,
                "dateEnd": date_end,
            },
            {"lang": "eng"},
        ],
        "$not": {
            "$or": [
                {"conceptUri": "http://en.wikipedia.org/wiki/Air_purifier"},              
            ]
        }
    },
}
events1 = event_registry.get_events_from_media_intelligence(query1, include_event_concepts=True)
print_titles(events1)
```

```python
thresholds1 = [
    {
        'uri': 'http://en.wikipedia.org/wiki/Air_quality_index',
        'score_threshold': 80
    },
]
filtered_events1 = filter_events_by_concept_thresholds(events1, thresholds1)
print_titles(filtered_events1)
```

```python
# trying to catch air pollution events connected to wildfire smoke
query2 = {
    "$query":{
        "$and": [
        {"$and": [
            {"conceptUri":"http://en.wikipedia.org/wiki/Air_pollution"},
            {"$or":[{"conceptUri":"http://en.wikipedia.org/wiki/Wildfire"},
                    {"conceptUri":"http://en.wikipedia.org/wiki/Bushfires_in_Australia"}
                   ]
            }
        ]
        },
        {"categoryUri":"dmoz/Science/Environment/Air_Quality"},
        {"dateStart": date_start,
         "dateEnd": date_end},
        {"lang": "eng"},
    ],
},
}
    
events2 = event_registry.get_events_from_media_intelligence(query2, include_event_concepts=True)
print_titles(events2)
```

```python
thresholds2 = [
    {
        'uri': 'http://en.wikipedia.org/wiki/Air_pollution',
        'score_threshold': 90
    },
    {
        'uri': 'http://en.wikipedia.org/wiki/Wildfire',
        'score_threshold': 70,
        'allow_not_exist': True
    }
]
filtered_events2 = filter_events_by_concept_thresholds(events2, thresholds2)
print_titles(filtered_events2)
```

```python
# trying to catch air pollution events connected to sand/dust storms
query3 = {
    "$query": {
        "$and": [
            {"conceptUri": "http://en.wikipedia.org/wiki/Dust_storm"},
            {"conceptUri": "http://en.wikipedia.org/wiki/Air_pollution"},
            {
                "dateStart": date_start,
                "dateEnd": date_end,
            },
            {"lang": "eng"},
        ],
        "$not": {
            "$or": [
                {"categoryUri": "dmoz/Shopping"},              
            ]
        },
    },
}
events3 = event_registry.get_events_from_media_intelligence(query3, include_event_concepts=True)
print_titles(events3)
```
